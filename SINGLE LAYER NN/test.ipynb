PROCEDURE

1 > SHAPE OF THE IMAGE IS 28*28 PIXELS
    SO TOTAL PIXELS WILL BE  = 784.
    
2  > GET THE WEIGHT FOR EACH PIXEL AND THEN CONVERT 2D IMAGE VALUES INTO SINGLE ARRAY 



```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
# SCALE THE VALUES
X_train =X_train / 255
X_test = X_test / 255
```

```python
# FLATTEN THE 2D IMAGE TO SINGLE ARRAT IMAGE
X_train_flattened = X_train.reshape(len(X_train),28*28)
X_test_flattened = X_test.reshape(len(X_test),28*28)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened.shape
```

```python
X_train_flattened[0]
```

```python
# KERAS - LIBRARY API 
# SEQUENTIA - LAYERS OF STACK
# DENSE - INPUT NEURON CONNECTED TO EVERYOTHER NEURON
# 10 - OUTPUT
# 784 - INPUT
# ACTIVATION FUNCTION SIGMOID
# ADAM OPTIMIZER - FAMOUS - FOR TRAINING NN
# LOSS FUNCTION - sparse_categorical_crossentropy - COMMONLY USED WITH INTEGER LABELS
# EPOCHS - ONE EPOCHS REFERS COMPLETE PASS THROUGH THE ENTIRE DATRASET DURING THE TRAINING PROCESS

model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flattened, y_train, epochs=10)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
# Y PREDICTED IS AN ARRAY
y_predicted = model.predict(X_test_flattened)
```
